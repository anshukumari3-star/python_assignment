```python
top_jobs = df['JobTitle'].value_counts().head(5)

plt.figure(figsize=(8,5))
sn.barplot(x=top_jobs.values, y=top_jobs.index, palette="viridis")
plt.title("Top 5 Most Common Jobs")
plt.xlabel("Count")
plt.ylabel("Job Title")
plt.show()
```
